Q3.Education and Income of Californian Adults

The State of California's Open Data Portal is a rich source of information about the lives of
Californians. A dataset on educational attainment and personal income among Californians over the years 2008 to 2014. The data are derived from the U.S. Census Current Population Survey. For each year, the table records the Population Count of Californians in many different combinations of age, gender, educational attainment, and personal income.
Data: educ_inc.csv
https://github.com/flaviovdf/evcomp2018/blob/master/data/educ_inc.csv
(a)	Analyse the data for the year 2014. (b) Consider adults (age >17)
â€¢	Each row of the table corresponds to a combination of age, gender, educational level, and income. There are 127 such combinations in all.
â€¢	As a first step focus on just one pair: educational attainment and personal income.
Requirement

```python
df3 = pd.read_csv("/content/educ_inc.csv")
df3
```

```python
df3['Year'] = pd.to_datetime(df3['Year'], format='%d-%m-%Y %H:%M')
```

```python
# (a)	Analyse the data for the year 2014. (b) Consider adults (age >17)
# â€¢	Each row of the table corresponds to a combination of age, gender, educational level, and income. There are 127 such combinations in all.
df4 = df3[
    (df3['Year'] >= '2014-01-01') &
    (df3['Year'] < '2015-01-01') &
    (df3['Age'] != '00 to 17')
]
df4
```

```python
# 1.Group the table by Educational Attainment and sum the Population Count in each category.
ans = df4.groupby('Educational Attainment')['Population Count'].sum()
ans
```

```python
# 2.Analyse the percentage distribution of educational attainment among adult Californians.
df_adult = df4[df4.Age != '00 to 17']
per = df_adult.groupby('Educational Attainment')['Population Count'].sum()
per = per.to_frame(name='Population Count')
per['Percentage'] = (per['Population Count']/per['Population Count'].sum())
per
```

```python
# 3.	Using pivot , get a contingency table (a table of counts) of adult Californians cross-classified
#  by Educational Attainment and Personal Income.
import numpy as np
pivot=df3.pivot_table(index=['Personal Income'],columns=['Educational Attainment'],aggfunc=np.sum)
pivot
```

```python
# 4.	Draw a bar chart to compare the personal income distributions of adult Californians who
# have no high diploma with those who have completed a Bachelor's degree or higher.
# (The difference in the distributions would be striking.
# There is a clear positive association between educational attainment and personal income).

df5 = df3[(df3['Age'] >= '18') & (df3['Personal Income']) & (df3['Educational Attainment'])]
df6 = df5[(df5['Educational Attainment']=='No high school diploma') ^ (df5['Educational Attainment']=="Bachelor's degree or higher")]
df6
pivot = df6.pivot_table(
    values='Population Count',
    index='Personal Income',
    columns='Educational Attainment',
    aggfunc='sum'
)
pivot

```

```python
for col in pivot.columns:
    pivot[col] = (pivot[col] / pivot[col].sum()) * 100
pivot.plot(
    y=["Bachelor's degree or higher", "No high school diploma"],
    kind="barh"
)
```
